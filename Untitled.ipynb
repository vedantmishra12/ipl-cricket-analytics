```python
import pandas as pd
import glob

files = glob.glob("ipl_data/*.csv")

df = pd.concat(
    [pd.read_csv(f, on_bad_lines='skip') for f in files],
    ignore_index=True
)

print(df.shape)
df.head()
 
```
